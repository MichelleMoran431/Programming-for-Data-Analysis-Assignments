Random number generation has applications in gambling , statistical sampling , computer stimulation and crytography and other areas where an unpredictable result is desirable such as security applications. Most computer programming languages include functions or library routines that provide random number generators. The syntax used to construct a new generator is numpy.random.default_rng().There is only one parameter called SEED used to initialize the BitGenerator and returns a generator. **#### 1. Normal Distribution[7,8,9]

This distribution is also known as Gaussian. Here is the probability density which is incorporated when creating a plot of results of the distribution where you can see the typical bell shape characteristic.









It fits the probability of many commonly occuring events  e.g Peoples height. Another real life example is seen in standardized testing where results can be made to have a normal distribution by either selecting the number and difficulty of questions or transferring the raw test scores into output scores by fitting them to a normal distribution.



The numpy.random.Generator.normal method is used. 

Its synax is : numpy.random.Generator.normal (loc=,scale=,size =)

The three parameters defined as :

loc- Mean ( " centre") of the distribution
scale - standard deviation- ("spread or width ")of the distribution
size- the shape of the output array

**Example : 

```python

from numpy import random
import numpy as np

mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.default_rng().normal(mu, sigma, 1000)
                                   
print(s)

#Visualise the samples and their probability density function in a histogram 

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()

```

#### 4. The Use of Seeds in Generating Pseudorandom Number [18,19,20,21, 22, 23,24]

Pseudorandom numbers are generated using BitGenerators and Generators (PRNG) which are algorithms that uses mathematical formulas to produce sequences of random numbers. These PRNGs start by using a seed state.The seed is a number/vector .It  does not need to be random, however if a pseudo number generator (PRNG) is reinitialized with the same seed, it will produce the same sequence of numbers. 

##### Generators

Objects that transform sequences of random bits from a BitGenerator into sequences of numbers that follow a specific distribution within a specific interval

##### BitGenerators

It provides a stream of random values. In order to generate reproducible streams, BitGenerators support setting their initial state via a seed. Seeds can be a arbitary-sized non-negative integer or list of integers. These inputs need to be processed in a high quality state which is done via the SeedSequence. This uses a hash techniques to ensure that even low quality seeds can generate high quality states. It uses a seed to initialize the BitGenerator which is then passed to a SeedSequence to derive the inital BitGenerator state. The most widely used general purpose PRNG.

In Numpy.random ( sampling) , there are two types of bitGenerators mentioned , Generator uses bits provided by PCG64 ( default) and RandomState ( legacy) uses MT19937.

**MT19937 : Mersenne Twister

This the most widely used general purpose PRNG. Its name is derived from the Mersenne Prime number. It implements the use of a 32- bit word length. 


**PCG64 :  Permuted Congruential Generator

This is relatively new PRNG, developed in 2014 and it applies a output permutation function. The state vector consists of 2 unsigned 128-bit values which are represented externally as python ints. One is the state of the PRNG which is advanced by the linear congruential generator (LCG) algorithm and other is a fixed odd increment used in the LCG.

Below is a diagram comparing both BitGenerators. It shows the catagories where the PCG64 outperforms the MT 19937 - for example its faster and has better statistical performance.





Looking at the performance timings such as the time in ns to product 1 random value from a specific distribution, the MT19937 is much slower as it requires 2 x 32-bit values to equal the output of the faster generators. 




**Importance of a good random Seed 

In the field of computer security , the choice of a good random seed is crucial. When a secret encrytion key is pseudorandomly generated , having the seed with allow one to obtain the key. These keys are used to protect data from unauthorized access over the internet.
Using the Seed function helps the optimization of codes where random numbers are used for testing and can be complex. Seed functions can be used to generate same random numbers again and again and this simplifies algorithm testing processes. 
If you want to reproduce the results for a particular run , you need to know what the current seed is. Pythons Random Generator doesnt store the seed in its memory.

##### Setting the seed  using the both Algorithms above : PCG64 and MT19937

Another difference between theses two algorithms is looking at setting/determining the seed value in order to get reproducible results. For both, the seed can be set using the random.seed() function for MT19937 and the default_rng for the PCG64. However to determine the seed , is easier using MT19937 whereas for the PCG64, the entropy state value is determined first.

In example 1 below ,It uses the function random.seed where its parameters are (a = seed value, version = 2 where a str, bytes or object gets converted to an int and all of its bits are used).The random module provides a fast pseudorandom number generator based on the Mersenne Twister algorithm, MT19937. 

```python
#Example 1

import random
import sys

# create a seed
seedValue = random.randrange(sys.maxsize)
# save this seed somewhere. So if you like the result you can use this seed to reproduce it

# Now, Seed the random number generator
random.seed(seedValue)
print("Seed was:", seedValue)

num = random.randint(10, 500)
print("Random Number", num)
```

Now using the above seed to generate the same results :

```python
import random

# Seed the random number generator with your value
random.seed(1593913663922940351)

num = random.randint(10, 500)
print("Random Number", num)
```

Example 2 shows how to use the PCG64 algorithm object- default_rng,to set the seed and reproduce results. Without passing any number through the object , the result will be random everytime you run it , however if you set the seed to e.g. 5 , then the result will be the same
